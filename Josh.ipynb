### Metric/Graphing

```python
import pandas as pd
import geopandas as gpd
%matplotlib inline
import matplotlib.pyplot as plt\

projects_data = pd.read_csv("projects.csv")
city_projects = projects_data.groupby('neighborhood')['name'].count().reset_index()
most_projects = city_projects.loc[city_projects['name'].idxmax()]

fig, ax = plt.subplots(figsize=(22,18))
city_projects.plot(kind='bar', x='neighborhood', y='name', ax=ax, color='blue')
ax.annotate(f'Most Projects: {most_projects["neighborhood"]}', xy=(most_projects.name, most_projects['name']), xytext=(5,5), textcoords='offset points', backgroundcolor='white')
ax.set_title("Projects In Pittsburgh Neighborhoods", fontsize=20)
ax.set_xlabel("Neighborhood", fontsize=18)
ax.set_ylabel("Number of Projects", fontsize=18)

plt.show()
```
