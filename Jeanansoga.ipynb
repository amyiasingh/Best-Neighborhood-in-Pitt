### ***The Metric***

When going about choosing a data set to showcase the public art in Pittsburgh it was quite easy since there is a data set titled “public art in Pittsburgh.” Though it became more difficult when selecting which column was consistent with the amount of public art in pittsburgh. For example columns like id, artist name, title, data all which account for certain public art display but I found that on the columns title and id consisted of consistent data for all public art in Pittsburgh whilst other columns would have gaps or completely lack information for ten plus pieces.

I decided to do a bar graph to showcase which city had the most public art and additionally I wanted to add some sort of annotator to show which city “won” to make it earlier to identify which neighborhood was the best based on my metrics. The Bar graph below data's represents the amount of public art in a neighborhood,revealing the neighborhood with the largest amount of public art. 

```python
import pandas as pd
import geopandas as gpd
%matplotlib inline
import matplotlib.pyplot as plt

pitt_public_art = pd.read_csv("Pittsburgh_Public_Art.csv")
city_art = pitt_public_art.groupby('neighborhood')['title'].count().reset_index()

most_neighbor = city_art.loc[city_art['title'].idxmax()]


fig, ax = plt.subplots(figsize=(20,16))
city_art.plot(kind='bar', x='neighborhood', y='title', ax=ax, color='pink')

ax.annotate(f'Most Art Pieces: {most_neighbor["neighborhood"]}', xy=(most_neighbor.name, most_neighbor['title']), xytext=(5,5), textcoords='offset points', backgroundcolor='white')


ax.set_title("Public Art In Pittsburgh Neighborhoods", fontsize=20)
ax.set_xlabel("Neighborhood", fontsize=18)
ax.set_ylabel("Amount Of Public Art", fontsize=18)

plt.show()
```

### **Code Explaination**

First I imported Pandas and matplotlib for data anylis and visualsation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Then I needed to grab, and read the data set "Public art in Pittsburgh" to follow through on my metrics.
additionally I needed to group by the nieghborhood in pittsburgh then count the column 'title' which acounts for all the public art in Pittsburgh.

```python
pitt_public_art = pd.read_csv("Pittsburgh_Public_Art.csv")
city_art = pitt_public_art.groupby('neighborhood')['title'].count().reset_index()
```

Then I accounted for the part of the program that would annotate the best neighborhood by whichever has the most amount of public art using the idxmax() function. then annoating the city with the most public art. 

```python
most_neighbor = city_art.loc[city_art['title'].idxmax()]
ax.annotate(f'Most Art Pieces: {most_neighbor["neighborhood"]}', xy=(most_neighbor.name, most_neighbor['title']), xytext=(5,5), textcoords='offset points', backgroundcolor='white'
```
